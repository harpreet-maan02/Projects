```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\harpr\OneDrive\Desktop\DSEU\Python Projects\Datasets\Diwali Sales Data.csv")
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.columns
```

```python
a=sns.countplot(x=df["Gender"])
for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_sum=df.groupby(["Gender"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sum_bar=sns.barplot(x='Gender',y='Amount',data=sales_sum)
for bars in sum_bar.containers:
    sum_bar.bar_label(bars)
```

```python
#women has more purchasing power
```

```python
ac=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ac.containers:
    ac.bar_label(bars)
```

```python
sales_ag=df.groupby(["Age Group"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
ac_bar=sns.barplot(x='Age Group',y='Amount',data=sales_ag)
# for bars in ac_bar.containers:
#     ac_bar.bar_label(bars)
```

```python
sales_state=df.groupby(["State"],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
ss_bar=sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
sales_oc=df.groupby(["Occupation"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
ac_bar=sns.barplot(x='Occupation',y='Amount',data=sales_oc)
plt.show()
```

```python
sales_pc=df.groupby(["Product_Category"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```
